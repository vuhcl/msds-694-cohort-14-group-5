```python
import os
import glob
import pandas as pd

pattern = os.path.join("./data/user_ratings/", "*.csv")
user_ratings = glob.glob(pattern)
pattern = os.path.join("./data/critic_ratings/", "*.csv")
critic_ratings = glob.glob(pattern)
pattern = os.path.join("./data", "*.csv")
album_info = glob.glob(pattern)
```

```python
user_ratings, critic_ratings, album_info
```

```python
dfs = []
for file in user_ratings:
    df = pd.read_csv(file)
    dfs.append(df)
user_df = pd.concat(dfs, ignore_index=True)
```

```python
dfs = []
for file in critic_ratings:
    df = pd.read_csv(file)
    dfs.append(df)
critic_df = pd.concat(dfs, ignore_index=True)
critic_df["snippet"] = critic_df["snippet"].apply(
    lambda x: x.strip() if type(x) == str else x
)
```

```python
dfs = []
for file in album_info:
    df = pd.read_csv(file)
    dfs.append(df)
album_df = pd.concat(dfs, ignore_index=True)
mask = album_df["release_year"].str.endswith("Date")
album_df.loc[mask, "release_year"] = album_df.loc[mask, "release_date"].apply(
    lambda x: x.split(" ")[1]
)
album_df.loc[mask, "release_date"] = album_df.loc[mask, "release_date"].apply(
    lambda x: x.split(" ")[0]
)
```

```python
user_df.to_csv("user_ratings.csv", index=False)
critic_df.to_csv("critic_ratings.csv", index=False)
album_df.to_csv("album_info.csv", index=False)
```
